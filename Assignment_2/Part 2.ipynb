```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
mnames = sc.textFile("/FileStore/tables/movie_metadata.tsv")
movieNames = mnames.map(lambda x :(x.split("\t")[0],x.split("\t")[2])).collect()
movies = {}
for i in movieNames:
      movies[i[0]] = i[1]
```

```python
import math
import re
from collections import Counter
import pandas as pd
import math
from pyspark.sql.functions import *

words = re.compile(r"\w+") 
movies_plots = sc.textFile("/FileStore/tables/plot_summaries.txt")
movies_plots = movies_plots.map(lambda x:x.split('\t'))
movies_plots.collect()
```

```python
inp = sc.textFile("/FileStore/tables/plot_summaries.txt")
inp = inp.map(lambda x:x.split('\t'))
map1 = inp.flatMap(lambda x: [((x[0],i.lower()),1) for i in x[1].split() if i not in stop_words and x[0] in movies.keys()])
map1.collect()
```

```python
reduction = map1.reduceByKey(lambda x,y:x+y)
reduction.collect()
```

```python
tf = reduction.map(lambda x: (x[0][1],(x[0][0],x[1])))
tf.collect()
```

```python
m1 = reduction.map(lambda x:(x[0][1],(x[0][0],x[1],1)))
m1.collect()
```

```python
m2 = m1.map(lambda x: (x[0],x[1][2]))
m2.collect()
```

```python
reduce2 = m2.reduceByKey(lambda x,y: x+y)
```

```python
n = len(inp.collect())
idf = reduce2.map(lambda x:(x[0],math.log10(n/x[1])))
rdd = tf.join(idf)
rdd1=rdd.map(lambda x: (x[1][0][0],(x[0],x[1][0][1],x[1][1],x[1][0][1]*x[1][1])))
rdd2=rdd1.map(lambda x:(x[0],x[1][0],x[1][1],x[1][2],x[1][3]))
df=rdd2.toDF(["DocumentID","Token","TF","IDF","TF_IDF"])
df.show()
```

```python
def cos(v1, v2):
    intersect = set(v1.keys()) & set(v2.keys())
    n = ([v1[x] * v2[x] for x in intersect])
    num = 0
    for i in n:
        num+=i
    s = ([v1[x] ** 2 for x in list(v1.keys())])
    s1 = 0
    for i in s:
        s1+=i
    s2 = ([v2[x] ** 2 for x in list(v2.keys())])
    sum2 = 0
    for i in s2:
        sum2+=i
    denom = math.sqrt(s1) * math.sqrt(sum2)
    if not denom:
        return 0.0
    else:
        return float(num) / denom 
def tv(text):
    words_return = words.findall(text)
    return Counter(words_return)
movies_plots = movies_plots.map(lambda x:(x[0],tv(str(x[1]))))
```

```python
user_Search = ['fun and comedy','violence','action and emotional','horror movies which are scientific']
user_Search = spark.sparkContext.parallelize(user_Search)
user_Search = user_Search.map(lambda x:x.lower()).filter(lambda x:len(x)>0).collect()
```

```python
for i in user_Search:
    print(">>>>>>>>>Search Text: "+i+"<<<<<<<<<<")
    if(len(i.split())<=1):    
        movieIds = df.filter(df["Token"]==i).sort(df["TF_IDF"].desc()).select(df["DocumentID"]).collect()
        for i in movieIds[:10]:
            print(movies[i[0]])
    else:
        v2 = tv(i)
        x = movies_plots.map(lambda x:(x[0],cos(x[1],v2))).sortBy(lambda x:-x[1]).collect()
        for i in x[:10]:
            print(movies[i[0]])
```
